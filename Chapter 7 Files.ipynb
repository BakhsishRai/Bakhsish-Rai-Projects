```python
# 3 Using with as a context manager - And reading play into Jupyter Notebook

# with is a context manager - good coding practice
# it allows us to work with a file within the block and automatically close it
# that means you don't have to close it

with open("romeojuliet.txt", "r") as r:
    # read the contents of the play into Jupyter Notebook
    play = r.read()
    print(play)
```

```python
#4 Removing white space - .strip() method

with open("romeojuliet.txt", "r") as r:

    for lines in r:
        lines = lines.strip()
        print(lines)  
```

```python
#5 Looping through play - .startswith() method

with open("romeojuliet.txt", "r") as r:
    
    # printing each line that starts with "ROMEO"
    for lines in r:
        if lines.startswith("ROMEO"):
            print(lines)         
```

```python
# 6 Count lines in play

with open("romeojuliet.txt", "r") as r:

    count = 0 # Why did we initialize count as 0?
    
    for lines in r:
        count = count + 1
    print("Line Count: ", count) # CLASS: Why is this print statement left justified?
```

```python
# 7 Counting lines in play that start with Romeo 

with open("romeojuliet.txt", "r") as r:

    count = 0
    romeo = 0

    
    for lines in r:
        if lines.startswith("ROMEO"):
            romeo = romeo + 1
        else:
            count = count + 1
            totalCount = count + romeo # Class: Why did I do this?

    print(f"""
    There are \x1B[1m{romeo}\x1B[0m lines that start with Romeo 
    and a total of \x1B[1m{totalCount}\x1B[0m lines in the play.""")
```

```python
# 8 Finding % of lines that start with Romeo 

with open("romeojuliet.txt", "r") as r:

    count = 0
    romeo = 0
    
    for lines in r:
        if lines.startswith("ROMEO"):
            romeo = romeo + 1
        else:
            count = count + 1
            totalCount = count + romeo # Class: Why did I do this?
            romeoPercent = round(((romeo/totalCount)*100),2) # Class: Why did I do this?
    
    print(f"""
    There are \x1B[1m{romeo}\x1B[0m lines that start with Romeo and a total of \x1B[1m{totalCount}\x1B[0m lines 
    in the play. That means lines with Romeo in them take about \x1B[1m{romeoPercent}%\x1B[0m of the play.""")
```

```python
#9 How to catch a user mistake when reading a file

file_name = input("Enter file name: ")

try:
    file_handle = open(file_name)
except:
    print(f"""File \x1B[1m{file_name}\x1B[0m cannot be opened.""")
```

The open function automatically opens in read "r" mode. You can also open it in write "w" mode and append "a" mode.

**Be careful with write mode! It will override anything in the file every time you run it!**
